```python
import numpy as np
import math
import statistics
from matplotlib import pyplot as plt
```

```python
num = list()
for x in range(0,1080):
    num.append(x)
print(len(num))
angle_value = [np.sin(2*math.pi*x/36) for x in num]
print(angle_value[0:36])
#週期是36
period = 36

#正確capturesize是18
capture_size = 18
timestep=capture_size 
phase_offset = 0

std = []
print(len(num)/period)
for i in range(0,18):
    value = []
    phase_offset = i
    for u in range(phase_offset, len(num),timestep):
        value.append(sum(angle_value[u:u+capture_size])/capture_size)
        #value = np.append(value,[angle_value[u:u+capture_size]])
    std.append(statistics.pstdev(value))
fig = plt.figure(figsize=(20,6))
plt.plot(std, label = "std")
plt.xlabel("phase_time_offset(no.)")
plt.ylabel("std")
plt.title(f"subject:    std phase")
plt.grid()
plt.xlim(0,len(std))
plt.ylim()
plt.legend()
plt.show
print(len(value))
```
